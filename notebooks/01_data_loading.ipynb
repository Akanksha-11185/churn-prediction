Customer Churn – Day 1 Exploration

This dataset contains customer-level information from a telecom company.
Each row represents a customer.
The target column is Churn, which shows whether a customer left the company.

* Columns Overview

tenure → how long the customer stayed

MonthlyCharges → how much they pay per month

Contract → monthly, yearly, etc

Churn → whether they left

* Initial Observations

Month-to-month customers churn more than long-term ones

Customers who churn tend to have higher monthly charges

Tenure appears to be strongly related to churn

These patterns will be used to build predictive models.
